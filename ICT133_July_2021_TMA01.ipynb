```python
'''
TMA01 Question 2(b)

Programming Steps:
1. Player keys in a minimum 4 up to 12 unique letters to form the target. ✔️
2. All letters will be converted to uppercase. ✔️
3. Valid letters are from A-Z. ✔️
4. Print Ammunition above the target string ✔️
5. Print Random Generated numbers next to the Ammunition ✔️
6. The random generated numbers must also be in capitaletters ✔️
7. The random generated numbers must form double the length of the target ✔️
8. Immediately provide next input for Hit 1. Enter to shoot: <Enter> ✔️
9. The random generated numbers also must be unique from the target and have 
   no duplicates ✔️
10. The number of chances to shoot will be the length of the target string. (Use a while loop) ✔️
11. So like if there is 4 target letters, there will be 4 attempts ✔️
12. Count the hit numbers ✔️
13. Program runs only 1 game ✔️
14. Now shoot at the target letter ✔️
15. Provide another random letter ✔️
10. Use a random letter from the ammunition letters ✔️
12. If the target letter is hit, record the hit ✔️
13. If the target letter is not hit, record miss ✔️
17. Display the hits ✔️
18. Display the misses ✔️
19. When a target is deleted, it will be removed from the print

Additional Steps:
1. 
2.
3.

'''

def main():
    import random   # random module imported
    
    special_characters = '!@#$%^&*()-+?_=,<>/'      # string of special characters
    numbers = '0123456789'                          # string of numbers
    empty_userinput = ''                            # empty string

    # Check unique characters function with target string parameter
    def check_unique_letters(target_string):
        # If at any time we encounter 2
        # same characters, return false
        for x in range(len(target_string)):                 # Get the length of user input
            for y in range(x + 1, len(target_string)):      # Iterate through this range to find variable y
                if target_string[x] == target_string[y]:    # index x and y in userinput would return False if equal
                    return False
        return target_string
        # If no duplicate characters
        # encountered, return true
    
    # Ammunition function with target string as parameter to get ammunition string
    def ammunition(target_string):
        letters = "abcdefghijklmnopqrstuvwxyz"          # string of letters
        Ammunition = target_string                 
        for i in range(len(target_string)):             # for loop to get target_string length
            while True:                                 # while loop to generate random ammunition letters multiple times
                Ammunition = target_string + letters[random.randint(1,25)].upper()  # generate random letters method
                if check_unique_letters(Ammunition):    # Check if random ammunition letters are unique
                    break
            target_string = Ammunition
        return target_string.upper()               

    # The Logic: If letter is in target length print Hit, else print Miss
    # Generate random letters function
    def random_letters():
        letters = 'abcdefghijklmnopqrstuvwxyz'                 # string of letters
        length_of_string = 1
        for i in range(length_of_string):                      # iterate through string length 
            random_letter = random.choice(letters).upper()     # generate 1 random letter from letters     
        return random_letter

    # hit_or_miss
    # Hit or Miss function to record if shot has hit or miss the target
    def hit_or_miss(target_string):
        for i in range(0, len(target_string)):                      # iterate through the range of 0 to target string length number    
            hit_target = 0
            ammo_position = random.choice(target_string).upper()    # random generated ammunition position
            if ammo_position in target_string:                      # validate if ammunition position is in target string
                target_hit = 'Target Hit'                           # target is hit if ammunition position is in target string 
                return target_hit # This is correct
            else:
                miss_target = 'Target Miss'
                hit_target += 1                                     # value of hit_target is appended by 1
                if hit_target == len(target_string):                
                    print('All Hit')
                else:
                    print(f'{hit_target} hits, {len(target_string) - hit_target} letters left')
                return miss_target

    # Run an infinite while loop that will stop when there is a break point
    while True:
        # Get target userinput and make all target userinput capital letters
        target_userinput = str(input('Enter Target Letters: ')).upper()
        
        if len(target_userinput) < 4:                                   # check target userinput length is less then 4
            print("String must be a minimum of 4 letters")
        elif len(target_userinput) > 12:                                # check target userinput length is greater then 12
            print("String must be a maximum of 12 letters")
        elif any(c in special_characters for c in target_userinput):    # check if target userinput has special characters
            print('No special characters allowed')
        elif any(n in numbers for n in target_userinput):               # check if target userinput has numbers
            print('No numbers allowed')
        elif any(i in empty_userinput for i in target_userinput):       # check if target userinput is a empty string
            print('Input must be at least 4 letters')
        elif target_userinput != check_unique_letters(target_userinput):    # check if target userinput has unique letters
            print('Must be unique letters')
        else:
            print('Ammunition: ', ammunition(target_userinput))
            print('Target String:', target_userinput)

            for i in range(len(target_userinput)):       # iterate through the target userinput length
                # print(i)
                i = i + 1                                                       # index variable i is appended by 1
                hit_input = str(input(f'Hit {i}. Press Enter to shoot: '))      # get hit input
                print(f'Letter {random_letters()} {hit_or_miss(target_userinput)}!')
                print(f'Target: {target_userinput}*')
                print(hit_input)
            else:
                length_of_random_letters = len(random_letters()) 
                print(f"{length_of_random_letters} hits, {length_of_random_letters} letters left")
                break

main()  




```
